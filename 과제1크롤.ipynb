```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

data=[]

page = 2 #페이지는 아래 주석처럼 입력을 받을수도 있으나 남용할 가능성을 고려해 2페이지만 하도록 설정
# int(input('몇 페이지까지 크롤링할래?'))
for i in range(1, page + 1): #레인지 뒤에꺼는 뒤에값 직전까지 만 출력이라 +1을 해주는것!!
    response = requests.get(f"https://kin.naver.com/search/list.naver?query=경주여행&page={i}")
    #쿼리(검색키워드)를 경주여행으로 고정, 페이지수를 f스트링으로 유동적으로 변경
    HTML = response.text
    soup = BeautifulSoup(HTML, 'html.parser')
    posts = soup.select(".basic1 > li") #나무태그 .basic1 > li를 아래 코드와 같이 포문으로 각각 돌면서 작업수행
    for post in posts:
        title = post.select_one('._searchListTitleAnchor').text
        link = post.select_one("._nclicks\\:kin\\.txt").attrs['href']
        date = post.select_one('.txt_inline').text
        category = post.select_one('.txt_g1._nclicks\\:kin\\.cat2').text
        review = post.select_one(".hit").text.split('답변수')[1].strip()
        #위는 각 카테고리별 태그를 셀렉원으로 텍스트 수집
        print(title,link,date,category,review)#테스트용 프린트 출력
        data.append([title,link,date,category,int(review)])#빈 data 리스트에 각각의 데이터를 추가
```

```python
#데이터 프레임
df = pd.DataFrame(data, columns=['제목','링크','날짜','카테고리','답변수'])#각 컬럼명을 지정하며 데이터프레임 생성
df
```

```python
#엑셀저장
df.to_excel('travel.xlsx')
```
